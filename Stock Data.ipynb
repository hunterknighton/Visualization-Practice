```python
import pandas as pd
import pandas_datareader.data as web# Package and modules for importing data; this code may change depending on pandas version
import datetime
```

```python
start = datetime.datetime(2016, 1, 7)
end = datetime.date.today()
stocks= ['GOOG', 'NVDA', 'AMZN','SPY']
```

```python
stockdata = web.DataReader('GOOG', 'yahoo', start, end)
```

```python
# create a data dictionary with a key for each stock so we can easily access that data
data = {}
i = 0
for stock in stocks:
    table = "Stock" + stock
    print("Downloading Data for",stock)
    data[stock] = web.DataReader(stock, 'yahoo', start, end)
    print(data[stock].head())
    
```

```python
def plot_stock(stock,col):
    figure = data[stock][col].plot(grid=True,title=stock)
    
    return figure
```

```python
i = 1
for stock in stocks:
    plt.figure(i)
    plot_stock(stock,'Close')
    i +=1
```
